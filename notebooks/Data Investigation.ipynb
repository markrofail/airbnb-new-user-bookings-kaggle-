```python
import pandas as pd
from pathlib import Path 
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.palettes import plasma
output_notebook()
%matplotlib inline
import matplotlib.pyplot as plt
from collections import Counter
```

```python
path_raw = Path("../data/raw")

for file in path_raw.iterdir():
        if file.is_file() and file.name.endswith("csv"):
           print(file.name)
```

```python
df = pd.read_csv(path_raw.joinpath('train_users_2.csv'))

cnt = Counter()
for record in df["country_destination"]:
    cnt[record] += 1
```

```python
# x and y axes
x_values = list(cnt.keys())
y_values = list(cnt.values())

p = figure(x_range=x_values, plot_height=250, title="Final Destination Count",
           toolbar_location=None, tools="")

p.vbar(x=x_values, top=y_values, width=0.9, 
       color=plasma(len(x_values)))

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
cnt = Counter()
for record in df["age"]:
    if record in range(1, 15):
        cnt['1, 15'] += 1
    elif record in range(15, 30):
        cnt['15, 30'] += 1
    elif record in range(30, 40):
        cnt['30, 40'] += 1
    elif record in range(40, 50):
        cnt['40, 50'] += 1
    elif record in range(50, 60):
        cnt['50, 60'] += 1
    elif record in range(60, 70):
        cnt['60, 70'] += 1
    elif record in range(70, 80):
        cnt['70, 80'] += 1
    elif record in range(80, 90):
        cnt['80, 90'] += 1
    elif record in range(90, 100):
        cnt['90, 100'] += 1
    elif record in range(100, 1000):
        cnt['120, 1000'] += 1
    elif record > 1000:
        cnt['> 1000'] += 1

print(cnt)
x_values = list(cnt.keys())
y_values = list(cnt.values())

p = figure(x_range=x_values, plot_height=250, title="Final Destination Count",
           toolbar_location=None, tools="")

p.vbar(x=x_values, top=y_values, width=0.9, 
       color='navy')

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```
